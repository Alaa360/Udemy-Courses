<h1> Udemy Courses analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('ggplot')
```

```python
df = pd.read_csv('udemyCourses.csv')
```

```python
df.head(3)
```

```python
df.columns
```

### All Subjects

```python
df.subject.unique()
```

### Subject with maximum number of courses

```python
max_courses = pd.DataFrame(df.['subject'].value_counts().sort_values(ascending=False))
max_courses
```

### Show all courses that are free of cost

```python
df[pd.to_numeric(df['price']) == 0]
```

### Show all courses that are paid

```python
df[df['is_paid'] == True]
```

### Top Selling Courses

```python
df.sort_values('num_subscribers', ascending=False)[:5]
```

### Top Least Selling Courses

```python
df.sort_values('num_subscribers', ascending=True)[:5]
```

### All graphic courses that are below 100

```python
df[(df.subject == 'Graphic Design') & (pd.to_numeric(df['price']) < 100)]
```

### All courses that are related to Python 

```python
df[df['course_title'].str.contains('Python')]
```

### All that are published in 2015

```python
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
df['Year'] = df['published_timestamp'].dt.year
df[df.Year == 2015]
```

### The max. number of subscribers per levels

```python
df.groupby('level')['num_subscribers'].max().sort_values(ascending = False)
```

```python
plt.bar(df['level'], df['num_subscribers'])
plt.xticks(rotation=45, size=8)
```
